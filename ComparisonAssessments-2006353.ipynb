```python
import subprocess
from BandShiftPlotCompare import DataCompObj,PlotComparisons
% matplotlib inline
```

```python
fList = ['Comp490_bric95.nc','Comp490_bric98.nc','Comp490_NN.nc']
tags = ['aph_bricaud95','aph_bricaud98','neural net']
homeDir = subprocess.Popen('echo $HOME',shell=True,stdout=subprocess.PIPE).communicate()[0][:-1]
relDir = 'DEBUG_BIOBANDSHIFT/TestSuite_2006353360/Comparisons_8D/'
fullDir = homeDir+'/'+relDir
paths = [fullDir+fList_i for fList_i in fList]
dataObjs=[DataCompObj(paths[i],tags[i]) for i in range(len(fList))]
colList=['#A62026','#348ABD','#23BC65']
```

```python
PlotComparisons(dataObjs,colList,plotTitle='Bandshifting  Comparison, $\lambda=490$')
```
